```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
```

```python
PATH = Service("C:/Program Files (x86)/Users/aaronhuang/Harp151")
driver= webdriver.Chrome(service=PATH)
driver.get("https://techwithtim.net")
# driver.close()#closes a tab
#driver.quiz()  closes the whole window

search = driver.find_element(By.NAME, "s")
search.clear()
search.send_keys("test")
search.send_keys(Keys.RETURN)

try:
    main = WebDriverWait(driver, 10).until(  
    EC.presence_of_element_located((By.ID, "main")) #locates main ID
    )
    print(main.text)
    
    articles = main.find_elements(By.TAG_NAME, "article") #locates article tag
    
    for i in articles:
        header = i.find_element(By.CLASS_NAME, "entry-summary") #locates entry-summary class 
        print(header.text)
finally:
        driver.quit()
```

```python
PATH = Service("C:/Program Files (x86)/Users/aaronhuang/Harp151")
driver= webdriver.Chrome(service=PATH)
driver.get("https://techwithtim.net")

#allows us to access the element from the text hyperlink v
link = driver.find_element("link text","Python Programming") # for updated code ->https://stackoverflow.com/questions/72754331/webdriver-object-has-no-attribute-find-element-by-link-text-selenium-scrip
link.click() #clicks the link


try: 
    element = WebDriverWait(driver, 10).until( #wait up to 10 seconders for the driver to find  element
        EC.presence_of_element_located((By.LINK_TEXT, "Beginner Python Tutorials")) #look for the hyperlink with the text "Beginner Python Tutorials"
    )
    element.clear() #clears so the keys we input so it doesn't append but replace what was there
    element.click()
    
    element = WebDriverWait(driver, 10).until( #wait up to 10 seconders for the driver to find ID
        EC.presence_of_element_located((By.ID, "sow-button-19310003")) #look for the ID called "sow-button-19310003"
    )
    element.click()
    
    driver.back()#goes to previous page
    driver.back()
    driver.back()
#   driver.forward() #goes to original page
    

except:  # if it does not load quit the site 
    driver.quit()
```

```python
csv_file = open("Selenium_cats.csv", "w", newline="", encoding= "utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["url"])

PATH = Service("C:/Program Files (x86)/Users/aaronhuang/Harp151")
driver= webdriver.Chrome(service=PATH)
driver.get("https://www.wikipedia.org/")

search = driver.find_element(By.ID, "searchInput")
search.clear()
search.send_keys("cat")
search.send_keys(Keys.RETURN)

try:
    main = WebDriverWait(driver, 10).until(  
    EC.presence_of_element_located((By.ID, "content")) #locates main ID
    )
    
    imgs = main.find_elements(By.TAG_NAME, "img") #find all img tags 
    for i in imgs:
        print(i.get_attribute("src"))
        url = i.get_attribute("src")
        csv_writer.writerow([url])

    
finally:
    driver.quit()
    
csv_file.close()
```
